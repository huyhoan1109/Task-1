```python
# Load package 
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
import numpy as np

from sklearn.datasets import load_iris

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

import warnings 
warnings.filterwarnings("ignore")
```

```python
# Load data
data = load_iris()
X_sepal = data['data'][:,0:2] # đài 
X_petal = data['data'][:,2:4] # cánh 
y = data['target'] # mục tiêu
```

```python
# Mô hình hóa theo đài và cánh 
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.scatter(X_sepal[:,0],X_sepal[:,1],c=y)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.subplot(1,2,2)
plt.scatter(X_petal[:,0],X_petal[:,1],c=y)
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.show()
```

```python
# chia vùng
def plot_regions(X, y, clf, test_idx=None, resolution=0.1):
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    color_maps = ListedColormap(colors[:len(np.unique(y))])
    x1_min = X[:,0].min() - 1
    x1_max = X[:,0].max() + 1
    x2_min = X[:,1].min() - 1
    x2_max = X[:,1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), np.arange(x2_min, x2_max, resolution))
    Z = clf.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contour(xx1, xx2, Z, alpha=0.4, cmap=color_maps)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    X_test, y_test = X[test_idx,:], y[test_idx]
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x = X[y == cl,0], y = X[y == cl,1], alpha=0.8, color=color_maps(idx), marker=markers[idx], label=cl)
```

# Train model with sepal 

```python
# cross validation và chuẩn hóa
X_sp_train, X_sp_test, y_train, y_test = train_test_split(X_sepal, y, test_size=0.2)

scaler = StandardScaler()
X_sp_train_std = scaler.fit_transform(X_sp_train)
X_sp_test_std = scaler.transform(X_sp_test)
X_combined_std = np.vstack((X_sp_train_std, X_sp_test_std))
y_combined = np.hstack((y_train, y_test))
print(f'Train {len(y_train)} samples')
print(f'Test {len(y_test)} samples')
```

```python
# Kernel tuyến tính
svm_sp_linear = SVC(C=1, kernel='linear')
folds = 4
i = 0
kf = KFold(n_splits=folds,shuffle=True)
plt.figure(figsize=(15,15))
for train_idx, val_idx in kf.split(X_sp_train_std):
    X_tr, X_val = X_sp_train_std[train_idx,:], X_sp_train_std[val_idx,:]
    y_tr, y_val = y_train[train_idx], y_train[val_idx]
    svm_sp_linear.fit(X_tr, y_tr)
    y_pred = svm_sp_linear.predict(X_val)
    score = accuracy_score(y_val, y_pred) * 100
    print(f'Fold {i+1} have an accuracy of {round(score,2)} %')
    i += 1
    plt.subplot(3,2,i)
    plt.subplots_adjust(hspace=0.4)
    plot_regions(X=X_sp_train_std, y=y_train, clf=svm_sp_linear, test_idx=val_idx)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
```

```python
# Kernel rbf
svm_sp_rbf = SVC(kernel='rbf', random_state=0)
i = 0
plt.figure(figsize=(15,15))
for train_idx, val_idx in kf.split(X_sp_train_std):
    X_tr, X_val = X_sp_train_std[train_idx,:], X_sp_train_std[val_idx,:]
    y_tr, y_val = y_train[train_idx], y_train[val_idx]
    svm_sp_rbf.fit(X_tr, y_tr)
    y_pred = svm_sp_rbf.predict(X_val)
    score = accuracy_score(y_val, y_pred) * 100
    print(f'Fold {i+1} have an accuracy of {round(score,2)} %')
    i += 1
    plt.subplot(3,2,i)
    plt.subplots_adjust(hspace=0.4)
    plot_regions(X=X_sp_train_std, y=y_train, clf=svm_sp_rbf, test_idx=val_idx)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
```
